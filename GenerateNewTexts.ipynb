```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data/train_cap2018.csv')
```

# Petits traitements

```python
import re
# pas du tout optimal
for i in range(dataset.shape[0]): 
  dataset.loc[i,'fulltext'] = dataset.loc[i,'fulltext'].replace('\r\n\r\n', '')
  dataset.loc[i,'fulltext'] = dataset.loc[i,'fulltext'].replace('      ', '')
  dataset.loc[i,'fulltext'] = re.sub(' +', ' ', dataset.loc[i,'fulltext'])
```

```python
file = open('data/allTextsC2.txt') #textes de niveau C2 nettoyés (pas de \r, \n, ainsi qu'un espace entre chaque mots)
temp = file.read()
temp = temp.replace('...', '[triplepoint]')
temp = temp.split('.')
for i in range(len(temp)):
  temp[i] = temp[i].replace('[triplepoint]', '...')
  temp[i] = re.sub('^ +', '', temp[i])
```

# Génération de nouveaux textes
Méthode de Markov : https://www.agiliq.com/blog/2009/06/generating-pseudo-random-text-with-markov-chains-u/

```python
meanNumberSentencesC2 = int(dataset.loc[0:49,'sentences'].mean(axis=0))
meanNumberWordsC2 = int(dataset.loc[0:49,'words'].mean(axis=0))
```

```python
# exécuter "pip3 install markovgen" dans terminal
import markovgen
def generateWithMarkov(file, number):
    file_ = open(file)
    markov = markovgen.Markov(file_)
    newTexts = []
    for i in range(number):
      newTexts.append(markov.generate_markov_text(max_size=meanNumberWordsC2))
    return newTexts
```

```python
import random
def generateSelectingRandomSentences(texts, number):
    newTexts = []
    for i in range(number):
      newText = []
      for j in range(meanNumberSentencesC2):
        newText.append(random.choice(texts))
      newTexts.append('. '.join(newText))
    return newTexts
```

```python
# suivant la méthode de markov
generateWithMarkov('data/allTextsC2.txt', 200)
```

```python
# suivant la deuxième méthode
generateSelectingRandomSentences(temp, 200)
```

```python
df = pd.DataFrame(generateSelectingRandomSentences(temp, 200))
del temp
df.columns = ['fulltext']
df['level1'] = 'C2'
dataset = dataset.append(df, ignore_index=True)
```
